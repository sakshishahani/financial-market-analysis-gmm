```python
import pandas as pd

energy = pd.read_excel("energy.xlsx")
marketcap = pd.read_excel("marketcap.xlsx")

energy.head()
```

```python
energy_long = energy.melt(id_vars=["Ticker","Name"],
                          var_name="Year",
                          value_name="Energy_Intensity")

energy_long.head()
```

```python
marketcap_long = marketcap.melt(id_vars=["Ticker","Name"],
                                var_name="Year",
                                value_name="Market_Cap")

marketcap_long.head()
```

```python
energy_long["Year"] = energy_long["Year"].astype(int)
marketcap_long["Year"] = marketcap_long["Year"].astype(int)
```

```python
data = pd.merge(energy_long, marketcap_long,
                on=["Ticker","Name","Year"])

data.head()
```

```python
data = data.sort_values(["Ticker","Year"])
data.head()
```

```python
data["mc_growth"] = data.groupby("Ticker")["Market_Cap"].pct_change()

data.head(10)
```

```python
data["Market_Cap"] = pd.to_numeric(data["Market_Cap"], errors="coerce")
data["Energy_Intensity"] = pd.to_numeric(data["Energy_Intensity"], errors="coerce")
```

```python
data["mc_growth"] = data.groupby("Ticker")["Market_Cap"].pct_change()
```

```python
data.head(10)
```

```python
data["ei_change"] = data.groupby("Ticker")["Energy_Intensity"].diff()
```

```python
data = data.dropna()
```

```python
import numpy as np

data["log_marketcap"] = np.log(data["Market_Cap"])

```

```python
data = data.dropna()
```

```python
import statsmodels.api as sm
```

```python
X = data[["ei_change", "log_marketcap"]]
y = data["mc_growth"]
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
data.to_csv("final_dataset.csv", index=False)
```
